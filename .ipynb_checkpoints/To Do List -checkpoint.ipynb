```python
to_do_list = []
while True:
    print("If you want to add in list type 'Add', if you want to delete in list type 'Delete' ")
    print("and if you want to clear screen type 'Clear' ")
    add_list=input()
    if add_list == "Add":
        print("What you want to add in list.")
        add_list = input()
        add_list = str(add_list)
        in_list  = [add_list]
        to_do_list.extend(in_list)
        print(to_do_list)
    elif add_list == "Clear":
        print("Clearing list ")
        to_do_list.clear()
        print(to_do_list)
    elif add_list == "Delete": 
        print("What do you want to Delete' ")
        add_list = input()
        to_do_list.remove(add_list)
        print(to_do_list)
```
